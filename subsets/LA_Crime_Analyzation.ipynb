```python
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
import time
import scipy as sp
import scipy.stats as stats
import numpy as np
import csv
%matplotlib inline

counter = 1;

fj=open("codesUnique.csv", "r")
codesReader = csv.reader(fj, delimiter = ',')
codes = []
for row in codesReader:
    codes.append(row)
    
def sepYear(year, data):
    ret = []
    for item in data:
        if(item[1] != 'Date Occured'):
            if(item[1][8:] == str(year)):
                ret.append(item)
    
    return ret

def graphHelper(data, counter, year):
        
    G = nx.Graph()
    G.add_node("AREA " + str(counter))
    for item in data:
        if(item[3] != 'Crime Code'):
            if(item[3] in G.nodes()):
                G["AREA " + str(counter)][item[3]]['weight'] += 1
            else:
                G.add_node(item[3])
                G.add_edge("AREA " + str(counter), item[3], weight=1)
    while(len(G.nodes) > 11):  
        min = 100000000000
        minNode = None
        for node in G.nodes:
            if(node != "AREA " + str(counter)):
                if(G["AREA " + str(counter)] [node] ['weight'] < min):
                    min = G["AREA " + str(counter)] [node] ['weight']
                    minNode = node
        G.remove_node(minNode)
    
    print("AREA " + str(counter) + '_'+ str(year) + ";Crime Code" + ";Frequency" + ";Description")
    for n in G.nodes:
        if(n != "AREA " + str(counter)):
            for i in codes:
                if i[1] == n:
                    desc = i[2]
                    print("AREA " + str(counter) + ";" + str(n) + ";" + str(G["AREA " + str(counter)][n]['weight']) + ";" + desc)
         
    print('\n')
    
    plt.figure(1, figsize = (5,5))
    nx.draw(G, with_labels=True, font_weight='bold')
    plt.savefig('labels_A' + str(counter) + '_' + year + '.png')
    plt.clf()
    
#     with open('tables_A' + str(counter) + '_' + year + '.csv', 'w') as csvfile:
#         fieldnames = ["AREA " + str(counter) + '_'+ str(year), "Crime Code", "Frequency", "Description"]
#         writer = csv.DictWriter(csvfile, lineterminator='\n', fieldnames=fieldnames)
#         writer.writeheader()
#         for n in G.nodes:
#             if(n != "AREA " + str(counter)):
#                 for i in codes:
#                     if i[1] == n:
#                         desc = i[2]
#                         writer.writerow({"AREA " + str(counter) + '_'+ str(year): "AREA " + str(counter), "Crime Code":str(n), "Frequency": str(G["AREA " + str(counter)][n]['weight']), "Description": desc})
        

def makeGraph(fi, counter):
    reader = csv.reader(fi, delimiter = ',')
    data = []
    data10 = []
    data11 = []
    data12 = []
    data13 = []
    data14 = []
    data15 = []
    data16 = []
    data17 = []

    for row in reader:
        data.append(row)
        
    data10 = sepYear(10, data)
    data11 = sepYear(11, data)
    data12 = sepYear(12, data)
    data13 = sepYear(13, data)
    data14 = sepYear(14, data)
    data15 = sepYear(15, data)
    data16 = sepYear(16, data)
    data17 = sepYear(17, data)
    
    
    graphHelper(data10, counter, '2010')
    graphHelper(data11, counter, '2011')
    graphHelper(data12, counter, '2012')
    graphHelper(data13, counter, '2013')
    graphHelper(data14, counter, '2014')
    graphHelper(data15, counter, '2015')
    graphHelper(data16, counter, '2016')
    graphHelper(data17, counter, '2017')
    
        #    print(desc)
    
def kendalTau(l1, l2):
    tau, p_value = stats.kendalltau(l1, l2)
    return tau


```
